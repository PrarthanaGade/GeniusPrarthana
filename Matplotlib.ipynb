```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style

ax = plt.axes(projection = "3d")

x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)

ax.scatter(x,y,z)
ax.set_title("3D Scattering")
plt.show()
```

```python
an = plt.axes(projection = "3d")

X = np.arange(-5,5,0.1)
Y = np.arange(-5,5,0.1)

X,Y = np.meshgrid(X,Y)
Z = np.sin(X) + np.cos(Y)

an.plot_surface(X,Y,Z, cmap = "Spectral")
an.set_title("3D plot")
an.set_xlabel("X")
an.set_ylabel("Y")
an.set_zlabel("Z")
an.view_init(azim = 0, elev = 90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

# Initialize the count for heads and tails
heads_tails = [0, 0]

# Set up the plot
fig, ax = plt.subplots()
bars = ax.bar(["Heads", "Tails"], heads_tails, color=["red", "blue"])

# Update function for the animation
def update(frame):
    heads_tails[random.randint(0, 1)] += 1
    bars[0].set_height(heads_tails[0])
    bars[1].set_height(heads_tails[1])
    ax.relim()
    ax.autoscale_view()

# Animate the plot
ani = plt.FuncAnimation(fig, update, frames=range(100000), repeat=False)
plt.show()
```
